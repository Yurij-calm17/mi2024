### **Питання екзаменаційних білетів з дисципліни МІАЗОВУ**  

#### **1. Використання баз даних**  
1. Структура реляційної бази даних для зберігання даних про бойові операції. Забезпечення доступу до інформації.  
2. Методи нормалізації даних у військових інформаційних системах. Вплив нормалізації на продуктивність бази даних.  
3. Використання NoSQL баз даних у військовому аналізі. Порівняння з реляційними базами даних.  
4. Реплікація бази даних у військових інформаційних системах. Підвищення надійності та доступності даних.  
5. Використання SQL-запитів для агрегованого аналізу військових даних. Створення аналітичних звітів.  
6. Організація резервного копіювання бази даних у військових інформаційно-аналітичних системах. Методи відновлення даних.  
7. Використання оператора SELECT для аналізу військових даних. Використання функцій та процедур.
---

#### **2. Використання методів машинного навчання**  
1. Основи використання методів машинного навчання у військовій аналітиці.  
2. Послідовність (етапи) розробки моделей машинного навчання у військових інформаційних системах.
3. Моделі прогнозування втрат військових підрозділів. Аналіз бойових даних.  
4. Використання нейромереж для інформаційно-анеалітичного забезпечення.  
5. Основи підготовки даних для машинного навчання у військових інформаційних системах.
6. Тренування моделей машинного навчання на реальних військових даних.
7. Основи оцінки точності моделей машинного навчання.

---

#### **3. Візуалізація результатів**  
1. Інтерактивні панелі для моніторингу бойової обстановки. Основні компоненти та принципи побудови.  
2. Використання бібліотек для створення інтерактивних звітів. Відображення військових даних.  
3. Основи розмітки гіпер-тексту для візуалізації військових даних.
4. Застосування стилів візуалізації військових даних.
5. Основи автоматизації поведінки веб-сторінки.
6. Створення інтерактивних візуалізацій військових даних.
7. Основи створення графічних візуалізацій військових даних.
