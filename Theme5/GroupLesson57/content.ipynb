## Розширений план-контент групового заняття  
**Тема 5. Заняття 7. Застосування принципів графічного дизайну для візуалізації даних**

---

### 1. Мета заняття

- **Ознайомити** слухачів із базовими принципами графічного дизайну, які є фундаментальними для створення ефективних візуалізацій даних.
- **Розкрити** методи та критерії застосування принципів дизайну (композиція, контраст, ієрархія, кольорова палітра, типографіка) для оптимізації сприйняття інформації.
- **Сприяти** розвитку навичок групової роботи через розбір кейсів та практичне застосування принципів графічного дизайну для створення прототипів інформаційних панелей.

### 3. Детальний план-конспект заняття

#### 3.1. Вступ (0–5 хв)
- Привітання учасників.
- Ознайомлення з темою заняття: "Застосування принципів графічного дизайну для візуалізації даних".
- Коротке пояснення, чому правильний вибір дизайнерських рішень критично важливий для ефективного сприйняття даних в умовах інформаційно-аналітичного забезпечення.

#### 3.2. Принципи графічного дизайну (5–25 хв)
- **Визначення:** Що таке графічний дизайн та його роль у створенні візуалізацій.
- **Основні принципи:**
  - **Композиція та баланс:** Як організувати елементи на сторінці, щоб забезпечити візуальну гармонію.
  - **Контраст та ієрархія:** Використання розмірів, кольорів і форм для виокремлення головних елементів.
  - **Кольорова палітра:** Як підбирати кольори для забезпечення чіткості та зручності сприйняття; приклади успішних та невдалих колірних рішень.
  - **Типографіка:** Роль шрифтів у підвищенні читабельності; оптимальні поєднання шрифтів для заголовків та текстового контенту.
  - **Whitespace (негативний простір):** Використання порожніх ділянок для покращення сприйняття та зменшення візуального перевантаження.
- **Аналіз прикладів:** 
  - Показати слайди або короткі відео з прикладами дашбордів і інфографіки, проаналізувати, що працює добре, а що ні.

#### 3.3. Використання графічного дизайну для візуалізації даних (25–40 хв)
- **Обговорення кейсів:**  
  - Приклади візуалізації даних у дашбордах: як елементи дизайну впливають на інтерпретацію статистичних даних.
  - Розгляд різних типів графіків (лінійні, стовпчикові, кругові) з точки зору дизайнерського підходу.
- **Критерії вибору візуальних елементів:**  
  - Відповідність типу даних (кількісні/якісні, часові, просторові).
  - Орієнтація на цільову аудиторію: які елементи допомагають швидко та ефективно донести інформацію до фахівців та ширшої аудиторії.
  - Забезпечення адаптивності: як дизайн змінюється для різних пристроїв.

#### 3.4. Групова робота (40–70 хв)
- **Завдання:**  
  - Кожна група отримує конкретний кейс (наприклад, "Візуалізація даних про чисельність військових підрозділів за регіонами" або "Аналіз трендів бюджетного фінансування").
  - Завдання полягає у розробці концепції візуалізації, що враховує всі основні принципи графічного дизайну: підбір кольорів, типографіка, композиція, використання whitespace, ієрархія інформації.
- **Процес роботи:**  
  - Групи обговорюють та розробляють коротку презентацію, у якій викладають свій підхід.
  - Викладач активно консультує, відповідає на питання та спрямовує дискусію для стимулювання критичного мислення.

#### 3.5. Презентація та обговорення (70–85 хв)
- **Презентація груп:**  
  - Кожна група має 3-4 хвилини, щоб представити свій прототип та пояснити аргументацію вибору візуальних елементів.
- **Обговорення:**  
  - Загальна дискусія, порівняння рішень груп, обговорення сильних і слабких сторін.
  - Відповіді на запитання та коментарі від викладача і однокурсників.

#### 3.6. Підсумки заняття (85–90 хв)
- **Підведення підсумків:**  
  - Узагальнення ключових принципів графічного дизайну та їхнього застосування для візуалізації даних.
  - Рекомендації щодо подальшого вивчення теми та практичного застосування отриманих знань.
- **Домашнє завдання:**  
  - Підготувати розширену презентацію або документ з аналізом можливих варіантів візуалізації даних для обраного кейсу, з обґрунтуванням дизайну та рекомендаціями щодо покращення.

Нижче наведено приклад коду, який демонструє застосування принципів графічного дизайну для візуалізації даних із використанням Chart.js для побудови графіків та HTML/CSS для оформлення інтерфейсу. Цей приклад ілюструє, як правильно підібрати кольорову палітру, типографіку, композицію та whitespace для створення сучасного дашборду.

```html
<!DOCTYPE html>
<html lang="uk">
<head>
  <meta charset="UTF-8">
  <title>Візуалізація даних з графічним дизайном</title>
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <!-- Підключення Chart.js -->
  <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
  <style>
    /* Основні стилі та принципи дизайну */
    body {
      font-family: 'Helvetica Neue', Arial, sans-serif;
      background-color: #f0f2f5;
      margin: 0;
      padding: 0;
      color: #333;
      line-height: 1.6;
    }
    header {
      background-color: #004080;
      padding: 20px;
      text-align: center;
      color: #fff;
    }
    nav {
      background-color: #0066cc;
      padding: 10px 0;
      text-align: center;
    }
    nav a {
      color: #fff;
      text-decoration: none;
      margin: 0 15px;
      font-weight: bold;
      transition: color 0.3s;
    }
    nav a:hover {
      color: #ffcc00;
    }
    main {
      padding: 20px;
    }
    /* Використання CSS Grid для адаптивного макету дашборду */
    .dashboard {
      display: grid;
      grid-template-columns: repeat(auto-fit, minmax(300px, 1fr));
      gap: 20px;
    }
    .card {
      background-color: #fff;
      border-radius: 8px;
      box-shadow: 0 4px 8px rgba(0, 0, 0, 0.1);
      padding: 20px;
      transition: transform 0.3s;
    }
    .card:hover {
      transform: translateY(-5px);
    }
    .card h2 {
      margin-top: 0;
      color: #004080;
      font-size: 1.5em;
      margin-bottom: 10px;
    }
    footer {
      background-color: #004080;
      color: #fff;
      text-align: center;
      padding: 10px;
      position: fixed;
      bottom: 0;
      width: 100%;
    }
  </style>
</head>
<body>
  <header>
    <h1>Візуалізація даних з графічним дизайном</h1>
    <p>Демонстрація використання принципів дизайну для покращення сприйняття інформації</p>
  </header>
  <nav>
    <a href="#">Головна</a>
    <a href="#">Дашборд</a>
    <a href="#">Аналітика</a>
    <a href="#">Контакти</a>
  </nav>
  <main>
    <section class="dashboard">
      <!-- Картка з графіком продажів -->
      <div class="card">
        <h2>Місячні продажі</h2>
        <canvas id="salesChart"></canvas>
      </div>
      <!-- Картка з круговою діаграмою розподілу категорій -->
      <div class="card">
        <h2>Розподіл категорій</h2>
        <canvas id="categoryChart"></canvas>
      </div>
    </section>
  </main>
  <footer>
    &copy; 2025 Інформаційна Аналітика
  </footer>
  
  <script>
    // Побудова стовпчикового графіку (Chart.js) для місячних продажів
    const salesCtx = document.getElementById('salesChart').getContext('2d');
    const salesChart = new Chart(salesCtx, {
      type: 'bar',
      data: {
        labels: ['Січень', 'Лютий', 'Березень', 'Квітень', 'Травень'],
        datasets: [{
          label: 'Продажі ($)',
          data: [15000, 18000, 17000, 19000, 21000],
          backgroundColor: 'rgba(54, 162, 235, 0.7)',
          borderColor: 'rgba(54, 162, 235, 1)',
          borderWidth: 1
        }]
      },
      options: {
        plugins: {
          title: {
            display: true,
            text: 'Місячні продажі'
          },
          legend: {
            display: false
          }
        },
        scales: {
          y: {
            beginAtZero: true,
            title: {
              display: true,
              text: 'Сума продажів'
            }
          }
        }
      }
    });
    
    // Побудова кругової діаграми для розподілу категорій продажів
    const categoryCtx = document.getElementById('categoryChart').getContext('2d');
    const categoryChart = new Chart(categoryCtx, {
      type: 'pie',
      data: {
        labels: ['Електроніка', 'Одяг', 'Книги', 'Інші'],
        datasets: [{
          label: 'Розподіл категорій',
          data: [40, 25, 20, 15],
          backgroundColor: [
            'rgba(255, 99, 132, 0.7)',
            'rgba(54, 162, 235, 0.7)',
            'rgba(255, 206, 86, 0.7)',
            'rgba(75, 192, 192, 0.7)'
          ]
        }]
      },
      options: {
        plugins: {
          title: {
            display: true,
            text: 'Розподіл продажів за категоріями'
          }
        }
      }
    });
  </script>
</body>
</html>
```

### Пояснення коду

1. **HTML-структура:**
   - Використовуються семантичні елементи: `<header>`, `<nav>`, `<main>` та `<footer>` для створення логічно структурованої сторінки.
   - Основний контент розташовується в секції з класом `dashboard`, яка організована за допомогою CSS Grid для адаптивності.
   - Кожна "картка" (`.card`) містить заголовок та елемент `<canvas>` для побудови графіків.

2. **CSS стилізація:**
   - Застосовано сучасні принципи дизайну: чистий фон, збалансовані відступи, використання контрастних кольорів (темно-синій, світло-блакитний) для виділення заголовків та даних.
   - Використано ефекти (тінь, плавна трансформація при наведенні) для створення відчуття простору та динамічності.
   - Адаптивний макет забезпечується через CSS Grid, що дозволяє компонентам правильно масштабуватися на різних пристроях.

3. **JavaScript:**
   - Використано Chart.js для створення інтерактивних графіків:
     - Стовпчиковий графік відображає місячні продажі.
     - Кругова діаграма демонструє розподіл продажів за категоріями.
   - Налаштування графіків включає заголовки, підписи осей та вибір кольорової палітри, що відповідає принципам контрасту та гармонії.

Цей приклад коду демонструє, як за допомогою правильно підібраних графічних елементів, адаптивного макету та інтерактивних компонентів створити сучасний дашборд, який ефективно відображає інформацію, забезпечуючи ясність, читабельність та естетичну привабливість.
