### 3. Детальний зміст лекції

#### 3.1. Основи графічного дизайну (5–25 хв)
- **Визначення графічного дизайну:**  
  Графічний дизайн – це мистецтво і наука створення візуального контенту, який передає інформацію та викликає емоційний відгук.
- **Основні принципи графічного дизайну:**
  - **Композиція та баланс:** Розташування елементів на сторінці для досягнення візуальної рівноваги.
  - **Контраст та ієрархія:** Використання розмірів, кольорів і форм для виокремлення головних елементів.
  - **Колірна палітра:** Вибір кольорів, що підкреслюють інформацію, забезпечують контрастність та гармонію.
  - **Типографіка:** Вибір шрифтів, їх розміри, відступи та вирівнювання для забезпечення читабельності.
  - **Використання простору (Whitespace):** Баланс між заповненими і порожніми зонами для підвищення зручності сприйняття.

- **Практичні приклади:**
  - Розбір візуальних кейсів: презентація слайдів із прикладами дашбордів, що мають сильний і слабкий дизайн, аналіз їх переваг і недоліків.

#### 3.2. Використання графічного дизайну для візуалізації даних (50–70 хв)
- **Підбір візуальних елементів:**  
  - Вибір типу діаграм та графіків залежно від характеру даних: лінійні графіки для трендів, стовпчикові для порівняння, кругові для часткових відношень.
  - Підбір інфографічних елементів для узагальнення складних даних.
- **Критерії адаптації дизайну до даних:**  
  - Забезпечення зрозумілості: елементи повинні бути простими та інтуїтивно зрозумілими.
  - Точність передачі даних: уникнення спотворень при візуальному представленні.
  - Адаптивність: дизайн повинен коректно відображатись на різних пристроях.
- **Практичний приклад:**  
  Обговорення кейсу: Як правильно підібрати графічні елементи для дашборду, що відображає дані про військову статистику (наприклад, чисельність підрозділів, бюджетні розподіли).

#### 3.3. Інтерактивна частина та групове обговорення (70–85 хв)
- **Групове завдання:**  
  - Кожна група отримує конкретний кейс, де потрібно розробити концепцію візуалізації даних з урахуванням принципів графічного дизайну.
  - Групи повинні визначити, які візуальні елементи (тип діаграми, колірна палітра, типографіка) найкраще підходять для заданих даних.
- **Обговорення:**  
  - Презентація рішень групами.
  - Колективний аналіз та обговорення критеріїв, аргументації вибору, обговорення альтернативних варіантів.

#### 3.4. Підсумки та висновки (85–90 хв)
- **Підведення підсумків:**  
  - Узагальнення основних принципів графічного дизайну та їхнього застосування у візуалізації даних.
  - Рекомендації щодо подальшого самостійного вивчення теми.
- **Відповіді на запитання:**  
  - Коротке Q&A, обговорення додаткових прикладів і роз’яснення незрозумілих моментів.

Нижче наведено приклад HTML-коду, який інтегрує бібліотеки Chart.js та Leaflet для побудови інтерактивного дашборду. У цьому прикладі на сторінці розміщено два розділи: один – для побудови стовпчикового графіку за допомогою Chart.js, а другий – для відображення карти з маркерами за допомогою Leaflet.

---

```html
<!DOCTYPE html>
<html lang="uk">
<head>
  <meta charset="UTF-8">
  <title>Інтерактивний дашборд: Chart.js + Leaflet</title>
  <!-- Підключення стилів для Leaflet -->
  <link rel="stylesheet" href="https://unpkg.com/leaflet/dist/leaflet.css" />
  <style>
    body {
      font-family: Arial, sans-serif;
      margin: 0;
      padding: 20px;
      background-color: #f4f4f4;
      color: #333;
    }
    header {
      text-align: center;
      margin-bottom: 20px;
    }
    .dashboard {
      display: flex;
      flex-wrap: wrap;
      gap: 20px;
      justify-content: space-around;
    }
    .card {
      background-color: #fff;
      border-radius: 8px;
      box-shadow: 0 2px 5px rgba(0, 0, 0, 0.1);
      padding: 20px;
      width: 45%;
    }
    /* Висота для карти */
    #map {
      height: 300px;
      width: 100%;
      border-radius: 8px;
    }
  </style>
</head>
<body>
  <header>
    <h1>Інтерактивний дашборд</h1>
    <p>Демонстрація використання Chart.js та Leaflet</p>
  </header>

  <div class="dashboard">
    <!-- Блок з графіком Chart.js -->
    <div class="card">
      <h2>Статистика відвідувань</h2>
      <canvas id="visitsChart"></canvas>
    </div>
    <!-- Блок з картою Leaflet -->
    <div class="card">
      <h2>Географія подій</h2>
      <div id="map"></div>
    </div>
  </div>

  <!-- Підключення Chart.js -->
  <script src="https://cdn.jsdelivr.net/npm/chart.js"></script>
  <!-- Підключення Leaflet JS -->
  <script src="https://unpkg.com/leaflet/dist/leaflet.js"></script>
  <script>
    // --- Chart.js: побудова стовпчикового графіку ---
    const ctx = document.getElementById('visitsChart').getContext('2d');
    const visitsChart = new Chart(ctx, {
      type: 'bar',
      data: {
        labels: ['Січень', 'Лютий', 'Березень', 'Квітень', 'Травень'],
        datasets: [{
          label: 'Відвідування сайту',
          data: [120, 150, 180, 130, 170],
          backgroundColor: 'rgba(54, 162, 235, 0.6)',
          borderColor: 'rgba(54, 162, 235, 1)',
          borderWidth: 1
        }]
      },
      options: {
        scales: {
          y: {
            beginAtZero: true,
            title: {
              display: true,
              text: 'Кількість'
            }
          }
        },
        plugins: {
          title: {
            display: true,
            text: 'Місячна статистика відвідувань'
          },
          legend: {
            display: false
          }
        }
      }
    });

    // --- Leaflet: ініціалізація карти та додавання маркерів ---
    // Створення карти та встановлення початкового центру
    const map = L.map('map').setView([50.4501, 30.5234], 10); // Координати Києва для прикладу

    // Додавання OpenStreetMap базового шару
    L.tileLayer('https://{s}.tile.openstreetmap.org/{z}/{x}/{y}.png', {
      attribution: '&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors'
    }).addTo(map);

    // Додавання маркерів з підписами
    const markersData = [
      { coords: [50.4501, 30.5234], popup: 'Центр Києва' },
      { coords: [50.464, 30.519], popup: 'Район Поділ' },
      { coords: [50.433, 30.521], popup: 'Бровари' }
    ];

    markersData.forEach(marker => {
      L.marker(marker.coords).addTo(map)
        .bindPopup(marker.popup)
        .openPopup();
    });
  </script>
</body>
</html>
```

### Пояснення:

1. **HTML-структура:**
   - Сторінка має заголовок та два основних блоки, розташованих у контейнері з класом `dashboard`.
   - Перший блок містить `<canvas id="visitsChart">` для побудови графіку за допомогою Chart.js.
   - Другий блок містить `<div id="map">`, де буде ініціалізовано карту за допомогою Leaflet.

2. **CSS стилізація:**
   - Визначені базові стилі для тексту, фонів та розташування елементів.
   - Використання Flexbox для розміщення блоків дашборду поруч один з одним.
   - Задані розміри для карти, щоб вона коректно відображалась.

3. **JavaScript:**
   - **Chart.js:** Створюється стовпчиковий графік, який показує статистику відвідувань за місяцями. Налаштовано кольори, підписи осей та заголовок графіку.
   - **Leaflet:** Ініціалізується карта з початковими координатами Києва, додається базовий шар з OpenStreetMap, а також маркери з інформаційними спливаючими вікнами (popup).

Цей приклад демонструє, як інтегрувати інструменти візуалізації даних для створення інтерактивного дашборду, який поєднує аналітичну графіку та геопросторове відображення даних.
